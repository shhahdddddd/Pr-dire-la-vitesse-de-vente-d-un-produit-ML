```python
import pandas as pd
df = pd.read_csv("produits_complet.csv")
df.head()
```

```python
df.info()
df.describe()
df.isna().sum()
```

```python
df = df.drop(columns=['Lien'])
```

```python
df.columns
```

```python
import numpy as np
df['Note'] = np.random.uniform(2.0, 8.0, len(df)).round(1)
df['Nb_avis'] = np.random.randint(0, 200, len(df))
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
duplicates = df['Nom'].duplicated().sum()
print(f"Number of duplicate product names: {duplicates}")
```

```python
df = df.groupby('Nom').agg({
    'Prix': 'mean',            # average price if different
    'Catégorie': 'first',      # assume same category
    'Note': 'mean',            # average rating
    'Nb_avis': 'sum',          # sum of reviews
    'En_stock': 'max',         # 1 if at least one in stock
    'Site': 'first'            # take first site
}).reset_index()

print(f"New dataset length after aggregation: {len(df)}")
```

```python
df['Nom'].duplicated().sum()   # should be 0 now
```

```python
df.dtypes
```

```python
df['Nb_avis'] = df['Nb_avis'].astype(int)
```

```python
df['En_stock'] = df['En_stock'].apply(lambda x: 1 if str(x).lower() == 'oui' else 0)
```
